```python
X = df[[
    'year',
    'country_encoded',
    'location_type_Rural',
    'location_type_Urban',
    'cellphone_access_encoded',
    'household_size',
    'gender_of_respondent_Female',
    'gender_of_respondent_Male',
    'relationship_with_head_encoded',
    'marital_status_encoded',
    'education_level_encoded',
    'job_type_encoded',
    'age_of_respondent'

]]
```

```python
y = df['bank_account_encoded']
```

```python
# output binar, task de clasificare
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier # diferite modele si metode de machine learning

from sklearn.model_selection import train_test_split # pentru splituirea datelor in date de antrenare si date de testare

from sklearn.metrics import accuracy_score, roc_auc_score
from sklearn.model_selection import cross_val_score # cross validation

from sklearn.model_selection import GridSearchCV # hyperparameter tuning

from sklearn.ensemble import VotingClassifier # se voteaza care model este mai bun

from imblearn.over_sampling import SMOTE # for unbalanced data, over-sampling the minority and under-sampling the majority class
from imblearn.under_sampling import RandomUnderSampler

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15)

# pentru balancing, in cazul la unbalanced dataset

smote = SMOTE()
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
```

```python
model = KNeighborsClassifier()
model.fit(X_train_smote, y_train_smote)
print(model.score(X_test, y_test))
```

```python
model = RandomForestClassifier()
model.fit(X_train, y_train) # smote este pentru unbalanced data
print(model.score(X_test, y_test))
```

```python
model = GradientBoostingClassifier()
model.fit(X_train, y_train)
print(model.score(X_test, y_test))
```

```python
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred)

print(f'Accuracy: {accuracy}')
print(f'ROC AUC: {roc_auc}')
```

```python
scores = cross_val_score(model, X_train, y_train, cv=5, scoring='roc_auc')
print(f'Cross-validated ROC AUC scores: {scores}')
```

```python
param_grid = {
    'n_estimators': [50, 100, 150],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
    # 'learning_rate': [0.01, 0.05, 0.1]
}

grid_search = GridSearchCV(GradientBoostingClassifier(), param_grid, cv=5, scoring='roc_auc')
grid_search.fit(X_train, y_train)

print(f'Best parameters: {grid_search.best_params_}')
print(f'Best score: {grid_search.best_score_}')

grid_search.score(X_test, y_test)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import GradientBoostingClassifier
from scipy.stats import uniform, randint

param_distributions = {
    'n_estimators': randint(50, 200),
    'max_depth': randint(5, 50),
    'min_samples_split': randint(2, 20),
    'learning_rate': uniform(0.01, 0.1),
    'subsample': uniform(0.6, 0.4),
    'max_features': ['auto', 'sqrt', 'log2']
}

randomized_search = RandomizedSearchCV(
    estimator=GradientBoostingClassifier(),
    param_distributions=param_distributions,
    n_iter=100,  # Number of combinations to try
    cv=5,
    scoring='roc_auc',
    n_jobs=-1,
    verbose=2,
    random_state=42
)

# Fit to the training data
randomized_search.fit(X_train, y_train)

# Output the best parameters
best_params = randomized_search.best_params_
print("Best Parameters:", best_params)

# Evaluate the best model on the test set
best_model = randomized_search.best_estimator_
test_score = best_model.score(X_test, y_test)
print("Test Score:", test_score)
```
