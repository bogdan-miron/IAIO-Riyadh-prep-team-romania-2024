```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/train.csv')
```

```python
pd.set_option('display.max_columns', None)
df
```

```python
# One hot encoding the sex feature
df = pd.get_dummies(df, columns = ['Sex'])
```

```python
df['Age_encoded'] = df['Age']
df.loc[df['Age'].isna(), 'Age_encoded'] = df['Age'].mean()
```

```python
df['Ticket'].value_counts()
```

```python
# Frequency encoding the 'tickets'
df['Ticket_encoded'] = df['Ticket'].map(df['Ticket'].value_counts())
```

```python
df
```

```python
df['Cabin'].value_counts()
```

```python
df['Cabin_encoded'] = df['Cabin'].map(df['Cabin'].value_counts())
```

```python
df.loc[df['Cabin_encoded'].isna(), 'Cabin_encoded'] = 0
```

```python
df.loc[df['Embarked'].isna(), 'Embarked'] = 'Q' # only 2 entries
```

```python
df.info()
```

```python
# One hot encoding the 'Embarked' feature
df = pd.get_dummies(df, columns = ['Embarked'])
```

```python
df
```

```python
df.info()
```

```python
X = df[[
    'Pclass',
    'Age_encoded',
    'SibSp',
    'Parch',
    'Fare',
    'Sex_female',
    'Sex_male',
    'Ticket_encoded',
    'Cabin_encoded',
    'Embarked_C',
    'Embarked_Q',
    'Embarked_S'
]]
```

```python
X.info()
```

```python
y = df['Survived']
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier # diferite modele si metode de machine learning
import xgboost as xgb
from xgboost import XGBClassifier
from sklearn.metrics import mean_absolute_error

from sklearn.model_selection import train_test_split # pentru splituirea datelor in date de antrenare si date de testare

from sklearn.metrics import accuracy_score, roc_auc_score
from sklearn.model_selection import cross_val_score # cross validation

from sklearn.model_selection import GridSearchCV # hyperparameter tuning

from sklearn.ensemble import VotingClassifier # se voteaza care model este mai bun

from imblearn.over_sampling import SMOTE # for unbalanced data, over-sampling the minority and under-sampling the majority class
from imblearn.under_sampling import RandomUnderSampler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15)

# for the unbalanced dataset
smote = SMOTE()
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
```

```python
model = KNeighborsClassifier()
model.fit(X_train, y_train)
y_preds = model.predict(X_test)
mae = mean_absolute_error(y_preds, y_test)
# print('MAE: ', mae)
print(model.score(X_test, y_test))
```

```python
model = RandomForestClassifier()
model.fit(X_train_smote, y_train_smote) # smote este pentru unbalanced data
print(model.score(X_test, y_test))
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
print(model.score(X_test, y_test))
```

```python
model = GradientBoostingClassifier()
model.fit(X_train, y_train)
print(model.score(X_test, y_test))
```

```python
model = SVC()
model.fit(X_train, y_train)
print(model.score(X_test, y_test))
```

```python
model = MLPClassifier()
model.fit(X_train, y_train)
print(model.score(X_test, y_test))
```

```python
param_grid = {
    'n_estimators': [100, 150 ,200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}
grid_search = GridSearchCV(RandomForestClassifier(), param_grid, cv=5, scoring='roc_auc')
grid_search.fit(X_train, y_train)

grid_search.score(X_test, y_test)
```

```python
grid_search.best_params_
```
