```python
# output binar, task de clasificare
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier # diferite modele si metode de machine learning

from sklearn.model_selection import train_test_split # pentru splituirea datelor in date de antrenare si date de testare

from sklearn.metrics import accuracy_score, roc_auc_score
from sklearn.model_selection import cross_val_score # cross validation

from sklearn.model_selection import GridSearchCV # hyperparameter tuning

from sklearn.ensemble import VotingClassifier # se voteaza care model este mai bun

from imblearn.over_sampling import SMOTE # for unbalanced data, over-sampling the minority and under-sampling the majority class
from imblearn.under_sampling import RandomUnderSampler

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15)

# pentru balancing, in cazul la unbalanced dataset

smote = SMOTE()
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
```

```python
model = KNeighborsClassifier()
model.fit(X_train, y_train)
print(model.score(X_test, y_test))
```

```python
model = RandomForestClassifier()
model.fit(X_train_smote, y_train_smote) # smote este pentru unbalanced data
print(model.score(X_test, y_test))
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
print(model.score(X_test, y_test))
```

```python
model = GradientBoostingClassifier()
model.fit(X_train, y_train)
print(model.score(X_test, y_test))
```

```python
model = SVC()
model.fit(X_train, y_train)
print(model.score(X_test, y_test))
```

```python
model = MLPClassifier()
model.fit(X_train, y_train)
print(model.score(X_test, y_test))
```

```python
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred)

print(f'Accuracy: {accuracy}')
print(f'ROC AUC: {roc_auc}')
```

```python
scores = cross_val_score(model, X_train, y_train, cv=5, scoring='roc_auc')
print(f'Cross-validated ROC AUC scores: {scores}')
```

```python
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}
grid_search = GridSearchCV(RandomForestClassifier(), param_grid, cv=5, scoring='roc_auc')
grid_search.fit(X_train, y_train)

grid_search.score(X_test, y_test)
```

```python
model1 = RandomForestClassifier()
model2 = DecisionTreeClassifier()
model3 = GradientBoostingClassifier()

voting_clf = VotingClassifier(estimators=[('rf', model1), ('dt', model2), ('gb', model3)], voting='soft')
voting_clf.fit(X_train, y_train)

voting_clf.score(X_test, y_test)
```

```python
# varianta de neural network

import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Check for NaN or infinity values in X_train
print(np.any(np.isnan(X_train)))  # Check for NaN values
print(np.any(np.isinf(X_train)))  # Check for infinity values

# Handle both NaN and infinity values
X_train = X_train.replace([np.inf, -np.inf], np.nan)  # Replace infinity with NaN
X_train = X_train.fillna(0)  # Replace NaN with 0

# Convert X_train and y_train to TensorFlow tensors, ensuring float32 dtype
X_train_tensor = tf.convert_to_tensor(X_train.values.astype(np.float32))  # Convert to float32
y_train_tensor = tf.convert_to_tensor(y_train.values.astype(np.float32))  # Convert to float32

# Convert X_test to a tensor, similar to X_train
X_test_tensor = tf.convert_to_tensor(X_test.values.astype(np.float32))

model = Sequential()
model.add(Dense(64, input_dim=X_train_tensor.shape[1], activation='relu'))
model.add(Dense(32, activation='relu')) # relu pentru activare hidden layer
model.add(Dense(1, activation='sigmoid')) # sigmoid pentru clasificare binara

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Use X_test_tensor for validation
model.fit(X_train_tensor, y_train_tensor, epochs=100, batch_size=32, validation_data=(X_test_tensor, y_test))
```
