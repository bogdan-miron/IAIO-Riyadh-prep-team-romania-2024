```python
import os
import shutil
import pandas as pd
```

```python
image_folder = 'C:\\Users\\retr0\\Downloads\\images\\images'
no_mask_folder = 'C:\\Users\\retr0\\Downloads\\no_mask'
mask_folder = 'C:\\Users\\retr0\\Downloads\\mask'
contest_data_folder = 'C:\\Users\\retr0\\Downloads\\contest_data'
csv_file = 'C:\\Users\\retr0\\Downloads\\train_labels.csv'
```

```python
os.makedirs(no_mask_folder, exist_ok=True)
os.makedirs(mask_folder, exist_ok=True)
os.makedirs(contest_data_folder, exist_ok=True)
```

```python
df = pd.read_csv(csv_file)
```

```python
labeled_images = set(df['image'])
```

```python
labeled_images
```

```python
for image_name in os.listdir(image_folder):
    image_path = os.path.join(image_folder, image_name)

    if not os.path.isfile(image_path):
        continue

    if image_name in labeled_images:
        label = df.loc[df['image'] == image_name, 'target'].values[0]

        if label == 0:
            dest_folder = no_mask_folder
        elif label == 1:
            dest_folder = mask_folder
        else:
            continue
    else:
        dest_folder = contest_data_folder
    shutil.move(image_path, os.path.join(dest_folder, image_name))
```
